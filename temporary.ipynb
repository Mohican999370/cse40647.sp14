```python
# This function can be used to select all columns that contain certain substrings
# Example usage: filter_index(df.columns,'nsusp') returns ['nsusp1', 'nsusp2', 'nsusp3', 'nsusp4', 'nsusp5', 'nsusp6', 'nsusp7']
def filter_index (indexes,keyword):
    filter_func = lambda s: keyword in s
    return [w for w in indexes if filter_func(w)]
```

## Load and parse data

```python
# Load the cleaned dataset
df = pd.read_csv('../MCPS_cleaned.csv',index_col=0,na_values='NA') #44 last col for 9th grade
```

```python
cuts = [14,26,36,46,56,66,80]
cuts2 = [0,14,26,36,46,56,66,80,87]
classifiers = ['LR','RF']
```

```python
# Select the dependent variable and drop data for 9-12th grades\
df2 = df.ix[:,0:cuts[0]]; df2['gender'] = df.male; df2['latentclass'] = df.attend_class
df2['graduated'] = df.nograd; df2['birth'] = df.birth
# Drop rows that have missing values because scikit-learn doesn't like that
df2 = df2.dropna()
#df2 = df2.fillna(-1)
print df2.shape
```

### Cummulative evaluation

```python
for i in range (0,len(cuts)):
    print 'Examining '+ str(i+6)+'th grade data:'
    # Select the dependent variable and drop data for 9-12th grades\
    df2 = df.ix[:,0:cuts[i]]; df2['gender'] = df.male; #df2['latentclass'] = df.attend_class
    df2.miss = df.ix[:,81+i]; df2['graduated'] = df.nograd; df2['birth'] = df.birth
    df2 = df2[df2.miss == 0]
    # Drop rows that have missing values because scikit-learn doesn't like that
    #df2 = df2.dropna()
    df2 = df2.fillna(-1)
    print 'Dataset shape after dropna(): ' + str(df2.shape)
    print 'Number of NOT graduated: ' + str(df2.graduated.sum())
    pm.runClassification(df2,'graduated',False,False,5,'topk',True,1.5,False,200,10,classifiers,.05)
    print '---------------------------------------------------------\n'
```

### Evaluating each year individually

```python
for i in range (0,len(cuts)-2):
    print 'Examining '+ str(i+6)+'th grade data:'
    # Select the dependent variable and drop data for 9-12th grades\
    df2 = df.ix[:,cuts[i]:cuts[i+1]]; df2['gender'] = df.male; #df2['latentclass'] = df.attend_class
    df2.miss = df.ix[:,81+i]; df2['graduated'] = df.nograd; df2['birth'] = df.birth
    df2 = df2[df2.miss == 0]
    # Drop rows that have missing values because scikit-learn doesn't like that
    #df2 = df2.dropna()
    df2 = df2.fillna(-1)
    print 'Dataset shape after dropna(): ' + str(df2.shape)
    print 'Number of NOT graduated: ' + str(df2.graduated.sum())
    pm.runClassification(df2,'graduated',True,False,5,'topk',True,1.5,False,200,10,classifiers,.1)
    print '---------------------------------------------------------\n'
    
```
